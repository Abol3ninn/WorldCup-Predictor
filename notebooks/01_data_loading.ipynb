### First I'll separate played vs unplayed matches

```python
played_matches = matches.dropna(subset=["home_score","away_score"])

future_matches = matches[matches["home_score"].isna()]
future_matches
```

```python
played_matches.shape
```

### Create the target column (0=Away Win, 1=Draw, 2=Home Win)

```python
def get_match_result(row):
    if row["home_score"] > row["away_score"]:
        return 2      # Home Win
    elif row["home_score"] < row["away_score"]:
        return 0      # Away Win
    else:
        return 1      # Draw


played_matches["target"] = played_matches.apply(get_match_result, axis=1)
```

```python
played_matches.head()
```

### Check class balance to know whether the dataset is balanced or not.

```python
played_matches["target"].value_counts()
```

```python
played_matches["target"].value_counts(normalize=True)
```

### Convert the date data into a proper datetime

```python
played_matches["date"] = pd.to_datetime(played_matches["date"])
```

### Sort the matches

```python
played_matches = played_matches.sort_values("date")
```

```python
played_matches = played_matches.reset_index(drop=True)
```
